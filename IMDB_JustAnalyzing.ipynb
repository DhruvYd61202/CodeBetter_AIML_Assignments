```python
import pandas as pd
```

```python
title_akas = pd.read_csv(
    'title.akas.tsv.gz',
    compression='gzip',
    sep='\t', 
    low_memory=False,  # prevent dtype warnings
    na_values='\\N' # IMDb uses \N for missing data
)
title_akas.head()
```

```python
title_ratings = pd.read_csv(
    'title.ratings.tsv.gz',
    compression='gzip',
    sep='\t', 
    low_memory=False,  # prevent dtype warnings
    na_values='\\N' # IMDb uses \N for missing data
)
title_ratings.head()
```

```python
title_basics = pd.read_csv(
    'title.basics.tsv.gz',
    compression='gzip',
    sep='\t',
    low_memory=False,
    na_values='\\N'
)
title_basics.head()
```

```python
akas_ratings = pd.merge(title_akas, title_ratings, left_on='titleId', right_on='tconst', how='inner')
# akas_ratings.shape
```

```python
akas_ratings.shape
akas_ratings.head()
```

```python
akas_ratings_merged = pd.DataFrame(akas_ratings[['title', 'averageRating']])
akas_ratings_merged.dropna(inplace=True)
print(akas_ratings_merged.shape)
# akas_ratings_merged.drop_duplicates(inplace=True)
print(akas_ratings_merged.shape)
akas_ratings_merged.isna().sum()
```
